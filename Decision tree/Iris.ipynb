```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
from sklearn.datasets import load_iris

dataset = load_iris()
```

```python
print(dataset['DESCR'])
```

```python
feature_name = dataset.feature_names
target_name = dataset.target_names
print(feature_name)
print(target_name)
```

```python
feature_data = dataset.data
```

```python
target_data = dataset.target
```

```python
feature_data
```

```python
target_data
```

```python
df = pd.DataFrame(feature_data,columns=feature_name)
```

```python
df.head()
```

```python
df['Output'] = target_data
```

```python
df.head()
```

```python
X = df.drop('Output',axis=1)
y = df['Output']
```

```python
from sklearn.model_selection import train_test_split 

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)

X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

classifier = DecisionTreeClassifier()

classifier.fit(X_train,y_train)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(classifier,filled=True)
```

```python
y_pred = classifier.predict(X_test)
```

```python
params= {
    'criterion':['gini', 'entropy', 'log_loss'],
    'splitter':['best', 'random'],
    'max_features':['sqrt', 'log2','auto'],
    'max_depth':[1,2,3,4,5]
}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(classifier,param_grid=params,cv=5,scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
y_pred = grid.predict(X_test)
print(y_pred)
```

```python
from  sklearn.metrics import accuracy_score,classification_report

report = classification_report(y_test,y_pred)
print(report)
```

```python
grid.best_score_
```

```python
classifier = DecisionTreeClassifier(criterion='entropy',max_depth=2,max_features='sqrt')
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score

score = accuracy_score(y_test,y_pred)
print("Accuracy Score :",score)
matrix = confusion_matrix(y_test,y_pred)
print(matrix)
```
