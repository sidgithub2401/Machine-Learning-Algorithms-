```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris

dataset = load_iris()
```

```python
print(dataset['DESCR'])
```

```python
feature_names = dataset.feature_names
output_features = dataset.target_names
```

```python
df = pd.DataFrame(dataset.data,columns=feature_names)
```

```python
df['Output'] = dataset.target
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
X= df.drop('Output',axis=1)
y = df['Output']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.10,random_state=42)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier()
classifier.fit(X_train_scaled,y_train)
```

```python
y_pred = classifier.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

accuracy = accuracy_score(y_pred,y_test)
report = classification_report(y_pred,y_test)
matrix = confusion_matrix(y_pred,y_test)

print("Accuracy :",accuracy)
print("Classification report :",report)
print("Confusion Matrix :\n",matrix)
```

```python
params={
    'criterion':['gini', 'entropy', 'log_loss'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5,6,7,8,9]
}
```

```python
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(classifier,cv=5,param_grid=params,scoring='accuracy')

grid.fit(X_train_scaled,y_train)
```

```python
grid.predict(X_test_scaled)
```

```python
grid.best_score_
```
