```python
X = df.drop('ProdTaken',axis=1)
y = df['ProdTaken']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30, random_state=42)
```

```python
cat_features = X.select_dtypes(include='object').columns
numeric_featues = X.select_dtypes(exclude='object').columns
```

```python
from sklearn.preprocessing import StandardScaler , OneHotEncoder
from sklearn.compose import ColumnTransformer

numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(drop='first')

preprocessor = ColumnTransformer(
    [
        ('OneHotEncoder',categorical_transformer,cat_features),
        ('Standard Scaler',numeric_transformer,numeric_featues)
    ]
)
```

```python
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,r2_score,precision_score,f1_score,recall_score
```

```python
models={
    'RandomForest': RandomForestClassifier()
}

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train,y_train)

    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    model_training_accuracy = accuracy_score(y_train,y_train_pred)
    model_train_f1 = f1_score(y_train,y_train_pred)
    model_precision_train_score = precision_score(y_train,y_train_pred)
    model_recall_train_score = recall_score(y_train,y_train_pred)

    
    model_testing_accuracy = accuracy_score(y_train,y_test_pred)
    model_test_f1 = f1_score(y_train,y_test_pred)
    model_precision_test_score = precision_score(y_train,y_test_pred)
    model_recall_test_score = recall_score(y_train,y_test_pred)


    print(list(models.keys()[i]))

    print("Model Performance for training Dataset")
    print("Accuracy: ",model_training_accuracy)
    print("F1 score: ",model_train_f1)
    print("precision_score:", model_precision_train_score)
    print("Recall Score :",model_recall_train_score)

    print()
    print("---------------------------------------------")
    print()

    print("Model Performance for Testing Data")
    print("Accuracy: ",model_testing_accuracy)
    print("F1 score: ",model_test_f1)
    print("precision_score:", model_precision_test_score)
    print("Recall Score :",model_recall_test_score)

    print()




```
