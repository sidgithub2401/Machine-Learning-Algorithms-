```python
X= df.drop('ProdTaken',axis=1)
y = df['ProdTaken']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
X_train
```

```python
cat_features = X.select_dtypes(include=object).columns
num_features = X.select_dtypes(exclude=object).columns
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer

numeric_transformer = StandardScaler()
category_transformer = OneHotEncoder(drop='first')

preprocessing =ColumnTransformer(
    [
        ('OneHotEncoder',category_transformer,cat_features),
        ('StandardScaler',numeric_transformer,num_features)
    ]
)
```

```python
X_train = preprocessing.fit_transform(X_train)
X_test = preprocessing.transform(X_test)
```

```python
pd.DataFrame(X_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,f1_score,recall_score,precision_score
```

```python
models ={
    "Decision Tree":DecisionTreeClassifier(),
    "Random Forest":RandomForestClassifier()
}

for i in range (len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train,y_train)

    y_test_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)

    train_score = accuracy_score(y_train,y_train_pred)
    train_model_precision_score = precision_score(y_train,y_train_pred)
    train_recall_score= recall_score(y_train,y_train_pred)

    
    test_score = accuracy_score(y_test,y_test_pred)
    test_model_precision_score = precision_score(y_test,y_test_pred)
    test_recall_score= recall_score(y_test,y_test_pred)

    print()
    print()
    print(list(models.keys())[i])

    print("Accuracy Score : ",train_score)
    print("Recall :",train_recall_score)
    print("Train model Precison :",train_model_precision_score)
    print()
    print("Accuracy Score : ",test_score)
    print("Recall :",test_recall_score)
    print("Test model Precison :",test_model_precision_score)






```

```python
rf ={
    'criterion':['gini', 'entropy', 'log_loss'],
    'n_estimators':[100,200,400,600,900,1000],
    'max_depth':[10,20,30,40],
    'max_features':['sqrt', 'log2']
}
```

```python
from sklearn.model_selection import RandomizedSearchCV

randomcvmodels=[
    ("RF",RandomForestClassifier(),rf)
]

```

```python
model_param={}
for name,model,params in randomcvmodels:
    randommodel = RandomizedSearchCV(estimator=model,param_distributions=rf,cv=3,n_iter=100,n_jobs=-1)
    randommodel.fit(X_train,y_train)

    model_param[name]= randommodel.best_params_

for model_name in model_param:
    print(model_name)
    print(model_param[name])
```
